### Imports

```python
from smdt import datasets, molecular_descriptors, classification, metrics
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Data

#### Reading Data

```python
USP = datasets.load_UspInhibition()
```

#### Calculating Descriptors

- Calculated descriptors separately on a remote Linux machine and saved it as a CSV
- Importing the file below (Note: File not on GitHub as file size is 98 MB)

```python
data = pd.read_csv('USP_All_Descriptors.csv')
```

#### Data Cleaning

```python
y = data['Target']
data.drop(['Target'],axis=1,inplace=True)
X = data
```

#### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

### Models

#### Gaussian Naive Bayes

```python
scores1, feature_names1, y_score1 = classification.fit_GaussianNB(X_train, X_test, y_train, y_test, 'active')
```

#### Linear SVC

```python
scores3, feature_names3, y_score3 = classification.fit_LinearSVC(X_train, X_test, y_train, y_test, 'active')
```

#### Random Forest Classifier

```python
scores4, feature_names4, y_score4 = classification.fit_RandomForestClassifier(X_train, X_test, y_train, y_test, 'active')
```

#### SGD Classifier

```python
scores5, feature_names5, y_score5 = classification.fit_SGDClassifier(X_train, X_test, y_train, y_test, 'active')
```

#### MLP Classifier

```python
scores6, feature_names6, y_score6 = classification.fit_MLPClassifier(X_train, X_test, y_train, y_test, 'active')
```

### Results

```python
fig = plt.figure(figsize=(15, 5))

plt.subplot(121)
metrics.plot_roc(y_test, y_score1, 'active', 'Gaussian NB')
metrics.plot_roc(y_test, y_score3, 'active', 'Linear SVC')
metrics.plot_roc(y_test, y_score4, 'active', 'Random Forest Classifier')
metrics.plot_roc(y_test, y_score5, 'active', 'SGD Classifier')
metrics.plot_roc(y_test, y_score6, 'active', 'MLP Classifier')

plt.subplot(122)
metrics.plot_prc(y_test, y_score1, 'active', 'Gaussian NB')
metrics.plot_prc(y_test, y_score3, 'active', 'Linear SVC')
metrics.plot_prc(y_test, y_score4, 'active', 'Random Forest Classifier')
metrics.plot_prc(y_test, y_score5, 'active', 'SGD Classifier')
metrics.plot_prc(y_test, y_score6, 'active', 'MLP Classifier')

plt.show()

pd.concat([scores1,scores3,scores4,scores5,scores6])
```
